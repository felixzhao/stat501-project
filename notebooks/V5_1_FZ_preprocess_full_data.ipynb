# preprocess

## fill nan to zero for registration types

```python
# List of columns to fill NaNs with zeros
cols_to_fill = ['ALCOHOL', 'BLOOD_PRESSURE', 'BLOOD_SUGAR', 'CIGARETTES', 'DIET_RULES', 'EXERCISE',
                'FASTING_BLOOD_SUGAR', 'LIFE_RULES', 'MEDICINE', 'MOOD', 'PAIN', 'SLEEP', 'STEPS', 'WAIST_HIP']

for col in cols_to_fill:
    df[col].fillna(0, inplace=True)
```

## one-hot encoding

```python
# One-Hot Encoding
df = pd.get_dummies(df, columns=['BMIClass'], prefix=['BMIClass'])
df = pd.get_dummies(df, columns=['closeReason'], prefix=['closeReason'])
df = pd.get_dummies(df, columns=['Intention.to.breastfeed_After.Birth'], prefix=['birth'])
```
